```python
from sklearn.impute import SimpleImputer
```

```python
import numpy as np
numeric_proessing = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
```

```python
numeric_proessing
```

```python
from sklearn.preprocessing import OneHotEncoder
categorical_processing = Pipeline([
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])  
categorical_processing
```

```python
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_proessing, ['gender','city']),  # assuming first three columns are numeric
        ('cat', categorical_processing,[ 'age','height'])])  # assuming last two columns are categorical
preprocessor
```

```python
from sklearn.pipeline import Pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])
pipeline
```
